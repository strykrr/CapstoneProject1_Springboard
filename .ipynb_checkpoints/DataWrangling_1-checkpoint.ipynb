```python
import pandas as pd
```

```python
german_data = pd.read_csv("C:\Springboard\Capstone1_Files\German_CreditRating_Data.csv")
```

```python
german_data.head()
```

```python
# Assigning column names 
german_data.columns = ['checking_account_status','duration','credit_history','purpose','credit_amount','savings_account','present_employment',
                 'installment_rate','personal_status_sex','debtors_guarontors','present_residence','property','age','installment_plans',
                  'housing','credits_at_bank','job','people_liable','telephone','worker','credibility']
```

```python
german_data.head()
```

```python
# Extracting numerical values from Categorical Columns

german_data.purpose = german_data.purpose.apply(lambda x:x[2])
german_data.checking_account_status = german_data.checking_account_status.apply(lambda x:x[2])
german_data.credit_history = german_data.credit_history.apply(lambda x:x[2])
german_data.savings_account = german_data.savings_account.apply(lambda x:x[2])
german_data.present_employment = german_data.present_employment.apply(lambda x:x[2])
german_data.personal_status_sex = german_data.personal_status_sex.apply(lambda x:x[2])
german_data.debtors_guarontors = german_data.debtors_guarontors.apply(lambda x:x[3])
german_data.property = german_data.property.apply(lambda x:x[3])
german_data.installment_plans = german_data.installment_plans.apply(lambda x:x[3])
german_data.housing = german_data.housing.apply(lambda x:x[3])
german_data.job = german_data.job.apply(lambda x:x[3])
german_data.telephone = german_data.telephone.apply(lambda x:x[3])
german_data.worker = german_data.worker.apply(lambda x:x[3])
```
