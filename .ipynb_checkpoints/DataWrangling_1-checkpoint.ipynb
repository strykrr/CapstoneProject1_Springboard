```python
import pandas as pd
```

```python
german_data = pd.read_csv("C:\Springboard\Capstone1_Files\German_CreditRating_Data.csv",header=None)
```

```python
german_data.info()
```

```python
# Assigning column names 
german_data.columns = ['checking_account_status','duration','credit_history','purpose','credit_amount','savings_account','present_employment',
                 'installment_rate','personal_status_sex','debtors_guarontors','present_residence','property','age','installment_plans',
                  'housing','credits_at_bank','job','people_liable','telephone','worker','credibility']
```

```python
german_data.head()
```

```python
categorical_cols= ['checking_account_status','credit_history','purpose','savings_account','present_employment',
                   'personal_status_sex','debtors_guarontors','property','installment_plans',
                  'housing','job','telephone','worker']
```

```python
#convert Categrorical Columns to Numerical
for col in categorical_cols:
    german_data[col] = german_data[col].apply(lambda x:x[-1])
```
