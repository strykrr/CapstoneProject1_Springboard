```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 
sns.set()
german_data = pd.read_csv("C:\Project_Files\Capstone1_Files\German_CreditRating_Cleaned_Data.csv")
```

```python
# Explore and Visualize each Atribute of the german creidt data  
# Attribute 1: (qualitative) 
# Status of existing checking account 
#A11 : ... < 0 DM 
#A12 : 0 <= ... < 200 DM 
#A13 : ... >= 200 DM / salary assignments for at least 1 year 
#A14 : no checking account 

good_accounts = pd.DataFrame(german_data['checking_account_status'][german_data['credibility']== 1])
bad_accounts = pd.DataFrame(german_data['checking_account_status'][german_data['credibility']== 0])
#print(good_status_accounts[['checking_account_status']]==2)
#initial histogram of good and bad accounts
plt.hist(good_accounts.checking_account_status,color='green')
plt.hist(bad_accounts.checking_account_status,color='blue')
plt.xlabel("Account types")
plt.ylabel("count of Account types")

# calculate percentage of good and bad accounts
good_accounts['account_type_percentages']= 1
bad_accounts['account_type_percentages'] = 1
good_acc_stats = good_accounts.groupby('checking_account_status').sum()*100/sum(good_accounts['account_type_percentages'])
bad_acc_stats = bad_accounts.groupby('checking_account_status').sum()*100/sum(bad_accounts['account_type_percentages'])
#plt.hist(bad_acc_percentages.account_type_percentages)
#print(good_acc_types)
#print((good_status_accounts)) 
print(good_acc_stats)
fig, ax = plt.subplots()
n_groups = 4
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8
 
rects1 = plt.bar(index, good_acc_stats.account_type_percentages, bar_width,
                 color='g',
                 label='Good')
 
rects2 = plt.bar(index + bar_width, bad_acc_stats.account_type_percentages, bar_width,
                 color='b',
                 label='Bad')
 
plt.xlabel('Account Types')
plt.ylabel('Percentages')
plt.title('Percentage of Account Types')
plt.xticks(index + bar_width, ('< 0 DM', '0 <= ... < 200 DM', '>= 200 DM', 'no checking account'))
plt.legend()
 
plt.tight_layout()
plt.show()
  
```

```python

# Attribute 2: (numerical) 
# Duration in month 

good_attr2 = pd.DataFrame(german_data['duration'][german_data['credibility']== 1])
bad_attr2 = pd.DataFrame(german_data['duration'][german_data['credibility']== 0])

#initial histogram of good and bad accounts
plt.hist(good_attr2.duration,bins=10,color='green') 
plt.hist(bad_attr2.duration,bins=10,color='blue')
plt.xlabel("Account types")
plt.ylabel("count of Account types")

plt.show()
  
```

```python
# Attribute 3: (qualitative) 
# Credit history 
# A30 : no credits taken/ all credits paid back duly 
# A31 : all credits at this bank paid back duly 
# A32 : existing credits paid back duly till now 
# A33 : delay in paying off in the past 
# A34 : critical account/ other credits existing (not at this bank

good_attr3 = pd.DataFrame(german_data['credit_history'][german_data['credibility']== 1])
bad_attr3 = pd.DataFrame(german_data['credit_history'][german_data['credibility']== 0])
#print(good_status_accounts[['checking_account_status']]==2)
#initial histogram of good and bad accounts
plt.hist(good_attr3.credit_history,bins=10,color='green') 
plt.hist(bad_attr3.credit_history,bins=10,color='blue')
plt.xlabel("Account types")
plt.ylabel("count of Account types")

# calculate percentage of good and bad accounts 
good_attr3['credit_history_percentages']= 1
bad_attr3['credit_history_percentages'] = 1
good_attr3_ptgs = good_attr3.groupby('credit_history').sum()*100/sum(good_attr3['credit_history_percentages'])
bad_attr3_ptgs = bad_attr3.groupby('credit_history').sum()*100/sum(bad_attr3['credit_history_percentages'])

fig, ax = plt.subplots()
n_groups = 5
index = np.arange(n_groups)
bar_width = 0.35
rects1 = plt.bar(index, good_attr3_ptgs.credit_history_percentages, bar_width,
                 color='g',label='Good')
rects2 = plt.bar(index + bar_width,bad_attr3_ptgs.credit_history_percentages, bar_width,
                 color='b',label='Bad')

plt.xlabel('Credit History')
plt.ylabel('Percentages')
plt.title('Percentage of Credit History')
plt.xticks(index + bar_width, ('no credits taken', 'bank credits paid duly', 'existing credits paid back duly', 'delay in payoff','other credits at banks'),rotation=45,ha='right')
plt.legend()
plt.tight_layout()   
plt.show()
```

```python
# function to filter persons with good and bad credit

def good_bad_credit_data(col1,col2='credibility'):
    good_credit_data = pd.DataFrame(german_data[col1][german_data[col2]== 1])
    bad_credit_data = pd.DataFrame(german_data[col1][german_data[col2]== 0])
    return (good_credit_data,bad_credit_data)

# function to calculate the percentage of each attribute
def percentages_good_bad(df1,df2,col1,col2):
    df1[col2]=1
    df2[col2]=1
    good_prcnt = df1.groupby(col1).sum()*100/sum(df1[col2])
    bad_prcnt = df2.groupby(col1).sum()*100/sum(df2[col2])
    return (good_prcnt,bad_prcnt)

# function for plotting histogram 
def hist_plot(df1,df2,col1,bin_num,labelx,labely):
    plt.hist(df1[col1],bins=bin_num,color='green') 
    plt.hist(df2[col1],bins=bin_num,color='blue')
    plt.xlabel(labelx)
    plt.ylabel(labely)     
        
# function to bar plot percentages of good and bad for an attribute
def bar_plot(df1,df2,col2,n_groups,labelx,labely,plt_ttl,xt_attr_types,degree=45,align='right'):
    fig, ax = plt.subplots()
    index = np.arange(n_groups)  
    bar_width = 0.20
    rects1 = plt.bar(index, df1[col2], bar_width,
                 color='g',label='Good')
    rects2 = plt.bar(index + bar_width,df2[col2], bar_width,
                 color='b',label='Bad')
    plt.xlabel(labelx)
    plt.ylabel(labely)
    plt.title(plt_ttl)
    plt.xticks(index + bar_width, xt_attr_types ,rotation=degree,ha=align)
    plt.legend()
    plt.tight_layout() 
    
```

```python
#Attribute 4: (qualitative) 
#Purpose 
#A40 : car (new) 
#A41 : car (used) 
#A42 : furniture/equipment 
#A43 : radio/television 
#A44 : domestic appliances 
#A45 : repairs 
#A46 : education 
#A47 : (vacation - does not exist?) 
#A48 : retraining 
#A49 : business 
#A410 : others 

good_attr4,bad_attr4 = good_bad_credit_data('purpose')
hist_plot(good_attr4,bad_attr4,'purpose',10,"purpose","total counts")
good_attr4_ptgs,bad_attr4_ptgs = percentages_good_bad(good_attr4,bad_attr4,'purpose','purpose_percentages')
xt_attr_types = ('car (new) ', 'car (used) ', 'furniture/equipment', 'radio/television ','domestic appliances','repairs', 
                  'education','(vacation - does not exist?)','retraining','business','others')
bar_plot(good_attr4_ptgs,bad_attr4_ptgs,'purpose_percentages',9,'purpose','percentages','percentage of Purposes',xt_attr_types,degree=45,align='right')
plt.show()
```
