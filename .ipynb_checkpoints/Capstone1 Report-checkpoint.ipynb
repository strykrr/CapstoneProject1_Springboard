### Statistical Analysis
* Chisquare Dependency test is performed on the Categorical Features of the datasets in determining the strength of association of each feature with Target.

* The Below listed Attributes have been ordered as per the strength of association of each attribute with the Target in classifying the customers.The Null Hypothesis is rejected when the significance level alpha is less than .05
    * Status of existing checking account 
    * Credit history 
    * Purpose
    * Property
    * Other installment plans
    * Housing
    * foreign worker 
    * Other debtors / guarantors
    * Savings account/bonds 
    * Personal status and sex
    * Present employment since
* The Below mentioned Attributes do not show promising dependency with the Target as the Null Hypothesis cannot be rejected based on the significance level.
    * Installment rate in percentage of disposable income
    * Present residence since 
    * Job 
    * Telephone
